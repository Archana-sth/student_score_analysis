```python
# Improting Libraries

#for data manipulation
import pandas as pd
import numpy as np

#for data visualization
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#I downloaded the "Students Exam Scores: Extended Dataset" from Kaggle, 
#which had a file size of 2.91 MB.
#I conducted data preprocessing by segregating the dataset into two different CSV files. 
#The first CSV file contains student information and the second CSV file contains the exam scores.

# read in my student information data
info = pd.read_csv("StudentInfo.csv")
print(info.head(3))
```

```python
# read in students score.
score = pd.read_csv("StudentScore.csv")
print(score.head(3))
```

```python
# merged two dataset into a new dataset.
student_info = pd.concat([info, score], axis = 1)
#student_info = pd.merge(info, score)
print(student_info.head(3))
```

```python
# check the size of dataset
print(student_info.shape)
```

```python
# Dataset information
student_info.describe()
```

```python
# count of null values
null_values = student_info.isnull().sum()
print(null_values)
```

```python
# Remove the "unnamed" column
student_info = student_info.drop("Unnamed: 0", axis = 1)
print(student_info.head(3))
```

```python
# change the column names
student_info.rename(columns={'Gender':'gender',
                   'EthnicGroup':'ethnic_group',
                   'ParentEduc':'parents_education',
                   'LunchType':'lunch_type',
                   'TestPrep':'test_preparation',
                   'ParentMaritalStatus':'parent_marital_status',
                   'PracticeSport':'practice_sports',
                   'IsFirstChild':'is_first_child',
                   'NrSiblings':'number_of_siblings',
                   'TransportMeans':'transport_means',
                   'WklyStudyHours':'weekly_study_hours',
                   'MathScore':'math_score',
                   'ReadingScore':'reading_score',
                   'WritingScore':'writing_score'}, inplace=True)
student_info.head(3)
```

```python
# change weekly study hours column
student_info["weekly_study_hours"] = student_info["weekly_study_hours"].str.replace("10-May","5-10")
student_info.head(3)
```

```python
# gender distribution
#creating a countplot and label the bars in the plot using the Seaborn and Matplotlib libraries
plt.figure(figsize = (5,3))

plt.title("Gender Distribution")

ax = sns.countplot(data=student_info, x= "gender")

# values of each gender
ax.bar_label(ax.containers[0])
             
plt.show()
```

from the above chart I have analyzed that the number of females in the data is more than number of males.

```python
group_by1 = student_info.groupby("test_preparation").agg({"math_score":"mean","reading_score":"mean", "writing_score":"mean"})
print(group_by1)

plt.figure(figsize= (5,3))
plt.title("Test Preparation")

ax = sns.countplot(data=student_info, x = "test_preparation")
plt.show()
```

An analysis of the chart above the average scores in Math, Reading, and Writing between students who completed test preparation (completed) and those who did not (none) have significant differences. Students who completed test preparation scored higher in all three subjects, The results suggest that completing test preparation may have a meaningful impact on student performance.

```python
# Parents education impacting student's score
group_by = student_info.groupby("parents_education").agg({"math_score":"mean","reading_score":"mean", "writing_score":"mean"})
print(group_by)

plt.figure(figsize = (5,3))
plt.title("Parent's Education")
sns.countplot(data=student_info, x= "parents_education")

plt.xticks(rotation=90)
plt.show()
```

students whose parents obtained Master's degree scored the highest in all three exams. There is a correlation between the exam scores of the students and their parent's highest education. This is because students with parents who have a degree scored higher than students who have parents without a degree

```python
#creating heatmap using seaborn to show the group_by data
sns.heatmap(group_by, annot = True)
print(group_by)
```

students with parents who hold associate's degree rank better than those with some high school.

```python
# creating the unique value of Ethnic Group
print(student_info["ethnic_group"].unique())
```

```python
# Distribution of Ethnic Group
EthnicGroupA = student_info.loc[(student_info["ethnic_group"] == "group A")].count()
EthnicGroupB = student_info.loc[(student_info["ethnic_group"] == "group B")].count()
EthnicGroupC = student_info.loc[(student_info["ethnic_group"] == "group C")].count()
EthnicGroupD = student_info.loc[(student_info["ethnic_group"] == "group D")].count()
EthnicGroupE = student_info.loc[(student_info["ethnic_group"] == "group E")].count()



EthnicGroup_list = [EthnicGroupA["ethnic_group"], EthnicGroupB["ethnic_group"], EthnicGroupC["ethnic_group"],
                  EthnicGroupD["ethnic_group"],EthnicGroupE["ethnic_group"]]

#labels for piechart
labels = ["Group A", "Group B", "Group C", "Group D", "Group E"]

plt.pie(EthnicGroup_list, labels=labels, autopct='%1.2f%%')# autopct shows percentage value

plt.title("Distribution of Students by Ethnic Group")

plt.axis('equal')  # Equal ensures that pie is drawn as a circle.
plt.show()
```

```python
# distribution of students from different ethnic groups and leveling the bars with counts.
axis = sns.countplot(data = student_info, x ='ethnic_group')

plt.title("Ethnic Group")

# Label all bars with their counts
axis.bar_label(axis.containers[0])
plt.show()

student_info.groupby(['ethnic_group'])[['math_score','writing_score','reading_score']].agg(['mean'])
```

students from Ethnic group E have the highest scores from the 3 subjects.
